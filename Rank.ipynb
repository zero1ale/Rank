```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import QuantileTransformer
%matplotlib inline
```

```python
df = pd.read_csv('pokemon.csv')
```

```python
aux = df[['species', 'attack', 'defense', 'spattack', 'spdefense', 'speed', 'total']].copy()
```

```python
aux.head()
```

```python
aux['default_rank_total'] = aux['total'].rank() ##Default
aux['max_rank_total'] = aux['total'].rank(method='max') ##máximo 
aux['min_rank_total'] = aux['total'].rank(method='min') ##minimo 
aux['avg_rank_total'] = aux['total'].rank(method='average') ##promedio 
aux['first_rank_total'] = aux['total'].rank(method='first') ##primer 
aux['dense_rank_total'] = aux['total'].rank(method='dense') ##densidad 
aux['NA_bottom_total'] = aux['total'].rank(na_option='bottom')
aux['pct_rank_total'] = aux['total'].rank(pct=True)
aux.sort_values('avg_rank_total').head(15)
```
