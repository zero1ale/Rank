```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
```

```python
df.head()
```

```python
aux = df[['Name','Publisher', 'Platform']]
```

```python
aux['default_rank_total'] = aux['Platform'].rank() ##Default
aux['max_rank_total'] = aux['Platform'].rank(method='max') ##m√°ximo 
aux['min_rank_total'] = aux['Platform'].rank(method='min') ##minimo 
aux['avg_rank_total'] = aux['Platform'].rank(method='average') ##promedio 

aux['dense_rank_total'] = aux['Platform'].rank(method='dense') ##densidad 
aux['NA_bottom_total'] = aux['Platform'].rank(na_option='bottom')
aux['pct_rank_total'] = aux['Platform'].rank(pct=True)
aux.sort_values('avg_rank_total').head(15)
```
